## Visualization

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

df = A01.copy()

fig = make_subplots(
    rows=1, cols=3,
    shared_yaxes=True,
    horizontal_spacing=0.05,
    subplot_titles=("GR", "RDEEP", "")
)

# -----------------------
# Track 1 - GR
# -----------------------
fig.add_trace(
    go.Scatter(x=df['GR'], y=df['DEPT'],
               mode='lines', name='GR',
               line=dict(color='green')),
    row=1, col=1
)

# -----------------------
# Track 2 - RDEEP
# -----------------------
fig.add_trace(
    go.Scatter(x=df['RDEEP'], y=df['DEPT'],
               mode='lines', name='RDEEP',
               line=dict(color='red')),
    row=1, col=2
)

# -----------------------
# Track 3 - RHOB (main x-axis of col=3)
# -----------------------
fig.add_trace(
    go.Scatter(x=df['RHOB'], y=df['DEPT'],
               mode='lines', name='RHOB',
               line=dict(color='orange')),
    row=1, col=3
)

# -----------------------
# Track 3 - NPHI (secondary axis overlay)
# -----------------------
fig.add_trace(
    go.Scatter(x=df['NPHI'], y=df['DEPT'],
               mode='lines', name='NPHI',
               line=dict(color='blue')),
    row=1, col=3
)

# Reverse depth
fig.update_yaxes(autorange="reversed")

# RHOB axis (bottom of track 3)
fig.update_xaxes(title_text="RHOB", range=[3, 1.5], row=1, col=3)

# Create overlay axis for NPHI
fig.layout["xaxis4"] = dict(
    title="NPHI",
    overlaying="x3",
    side="top",
    range=[0, 0.6]
)

# Assign NPHI trace to xaxis4
fig.data[3].update(xaxis="x4")

fig.update_layout(
    height=2000,
    width=1200,
    title="A01",
)

# GR range
fig.update_xaxes(title_text="GR", range=[30, 100], row=1, col=1)

# RDEEP range
fig.update_xaxes(
    type="log",
    title_text="RDEEP (log scale)",
    range=[-1,2],
    row=1, col=2
)


fig.show()
```

```python
df = A01.copy()
df['INV_SQRT_RDEEP'] = 1 / np.sqrt(df['RDEEP'])

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df['RHOB'],
        y=df['INV_SQRT_RDEEP'],
        mode='markers',
        marker=dict(size=5, color='blue'),
        name='Data Points'
    )
)

fig.update_layout(
    title="A01",
    xaxis_title="RHOB (g/cc)",
    yaxis_title="1 / √RDEEP",
    height=800,
    width=800
)

# Log scale on Y-axis
fig.update_xaxes(
    range=[3.0, 1.5]   # RHOB range
)

fig.update_yaxes(
    type="linear",
    range=[0, 1.0]     # 1/sqrt(RDEEP) range
)

fig.show()
```

### Density Neutron

```python
import numpy as np
import plotly.graph_objects as go

df = A01.copy()

fig = go.Figure()

# Scatter points (no depth coloring)
fig.add_trace(
    go.Scatter(
        x=df['NPHI'] * 100,   # convert to porosity units
        y=df['RHOB'],
        mode='markers',
        marker=dict(
            size=5
        ),
        name='Data'
    )
)

# ----- Matrix Trend Lines -----
rho_fl = 1.0  # fluid density

matrix = {
    "Sandstone": 2.65,
    "Limestone": 2.71,
    "Dolomite": 2.87
}

phi = np.linspace(0, 0.60, 100)

for name, rho_ma in matrix.items():
    rho = rho_ma * (1 - phi) + rho_fl * phi

    fig.add_trace(
        go.Scatter(
            x=phi * 100,
            y=rho,
            mode='lines',
            name=name
        )
    )

# Layout formatting
fig.update_layout(
    title="A01",
    xaxis_title="Neutron Porosity (p.u.)",
    yaxis_title="Bulk Density (g/cc)",
    height=800,
    width=800
)

# Reverse density axis (industry style)
fig.update_yaxes(range=[3.0, 1.5])
fig.update_xaxes(range=[0, 60])

fig.show()
```
